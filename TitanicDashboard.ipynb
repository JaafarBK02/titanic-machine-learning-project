```python
import pandas as pd

results = [
    {'Model': 'Logistic Regression', 'Accuracy': 0.78},
    {'Model': 'KNN', 'Accuracy': 0.676 },
    {'Model': 'SVM', 'Accuracy': 0.6369 },
    {'Model': 'Decision Tree', 'Accuracy': 0.7877 },
    {'Model': 'Random Forest', 'Accuracy': 0.8045},
    {'Model': 'Naive Bayes', 'Accuracy': 0.7598},
    {'Model': 'MLP', 'Accuracy': 0.7709},
    {'Model': 'XGBoost', 'Accuracy': 0.7989}
]

# Save results as CSV
df = pd.DataFrame(results)
df.to_csv("model_results.csv", index=False)

# Find the best model
best_model = max(results, key=lambda x: x['Accuracy'])

# Streamlit app code as a string
dashboard_code = f'''

import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# Title
st.title("Titanic - Classification Model Comparison Dashboard")

# Best Model
st.success("Best Model: **{best_model['Model']}** with **{best_model['Accuracy']*100:.2f}%** accuracy")

# Load results
df = pd.read_csv("model_results.csv")

# Show dataframe
st.subheader(" Model Comparison Table")
st.dataframe(df)

# Bar chart
st.subheader("Accuracy Comparison")
fig, ax = plt.subplots()
ax.bar(df['Model'], df['Accuracy'], color='skyblue')
ax.set_ylabel("Accuracy")
ax.set_ylim(0, 1)
plt.xticks(rotation=45)
st.pyplot(fig)
'''

# Save Streamlit app to a .py file
with open("titanic_dashboard.py", "w") as f:
    f.write(dashboard_code)
```
